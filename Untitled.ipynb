```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('QuebecProv.csv', sep=';',encoding='ANSI')
df['Population estimée'].replace(' ', '', inplace=True, regex=True)
df['Population estimée'] = df['Population estimée'].astype(int)
```

```python
def dms_to_latlon(x):
    if 'N' in x:
        x = x.replace('°', '').replace("'", "").replace('"', '').replace('N', '').replace('O', '').strip().split(' ')
        x = int(x[0]) + int(x[1])/60 + int(x[2])/3600
        return x
    elif 'O' in x:
        x = x.replace('°', '').replace("'", "").replace('"', '').replace('N', '').replace('O', '').strip().split(' ')
        x = int(x[0]) + int(x[1])/60 + int(x[2])/3600
        return -x
```

```python
df['lat'] = df.Latitude.apply(dms_to_latlon)
df['lon'] = df.Longitude.apply(dms_to_latlon)
```

```python
plot = df[:500].plot.scatter('lon', 'lat', c='xkcd:pale green',  s=df['Population estimée']/40, alpha = 0.5, figsize = (50,50))
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
ax.set_facecolor('xkcd:black')
ax.axes.get_yaxis().set_visible(False)
ax.axes.get_xaxis().set_visible(False)
plt.plot()
```

```python
df.sort_values(['Population estimée'], ascending=False, inplace=True)
```
